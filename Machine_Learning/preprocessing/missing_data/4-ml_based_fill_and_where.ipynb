```python
import seaborn as sns
import missingno as msno
import numpy as np

df = sns.load_dataset('titanic')
df = df.select_dtypes(include = ['float64',"int64"])

print(df.head())
df.isnull().sum(axis=0)
```

```python
ndf = np.array(df)

ndf.shape
```

```python
from ycimpute.imputer import knnimput
dff = knnimput.KNN(k = 4).complete(ndf)
```

```python
import pandas as pd
df_filled = pd.DataFrame(dff,columns = df.columns)
df_filled
```

```python
df_filled.isnull().sum()
```

```python
from ycimpute.imputer import EM
dff = EM().complete(np.array(df))

df3 = pd.DataFrame(dff, columns=df.columns)
df3
```

```python
df3.isnull().sum()
```
