```python
import pandas as pd
import numpy as np 

df = pd.read_csv("WHO COVID-19.csv")
df
```

```python
df_Europe = df[df["WHO Region"] == "Europe"]
df_Europe
```

```python
X = df_Europe.iloc[:, 2:10]
X
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled = scaler.fit_transform(X)

df_scaled = pd.DataFrame(scaled, columns = X.columns, index= df_Europe["Name"])
df_scaled
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

plt.figure(dpi=150)

linkage_matrix = linkage(df_scaled, method = "ward", metric = "euclidean")

dendrogram(linkage_matrix, labels = df_scaled.index.to_list())
plt.show()
```
