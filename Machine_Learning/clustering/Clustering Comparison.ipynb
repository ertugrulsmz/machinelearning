```python
from sklearn import datasets

circles = datasets.make_circles(n_samples = 1500, noise = 0.05, factor = 0.5)
X_circle, y_circle = circles

moons = datasets.make_moons(n_samples = 1500, noise = 0.05)
X_moon, y_moon = moons
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi = 150)

ax1 = plt.subplot(1,2,1)
ax1.scatter(X_circle[:,0], X_circle[:,1], c = y_circle)

ax2 = plt.subplot(1,2,2)
ax2.scatter(X_moon[:,0], X_moon[:,1], c = y_moon)
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 2)

kmeans.fit(X_circle)
kmeans_circle = kmeans.labels_

kmeans.fit(X_moon)
kmeans_moon = kmeans.labels_
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi = 150)

ax1 = plt.subplot(1,2,1)
ax1.scatter(X_circle[:,0], X_circle[:,1], c = kmeans_circle)

ax2 = plt.subplot(1,2,2)
ax2.scatter(X_moon[:,0], X_moon[:,1], c = kmeans_moon)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(n_clusters = 2, affinity = "euclidean", linkage = "single")

agg.fit(X_circle)
agg_circle = agg.labels_

agg.fit(X_moon)
agg_moon = agg.labels_
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi = 150)

ax1 = plt.subplot(1,2,1)
ax1.scatter(X_circle[:,0], X_circle[:,1], c = agg_circle)

ax2 = plt.subplot(1,2,2)
ax2.scatter(X_moon[:,0], X_moon[:,1], c = agg_moon)
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps = 0.15)

dbscan.fit(X_circle)
dbscan_circle = dbscan.labels_

dbscan.fit(X_moon)
dbscan_moon = dbscan.labels_
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi = 150)

ax1 = plt.subplot(1,2,1)
ax1.scatter(X_circle[:,0], X_circle[:,1], c = dbscan_circle)

ax2 = plt.subplot(1,2,2)
ax2.scatter(X_moon[:,0], X_moon[:,1], c = dbscan_moon)
plt.show()
```
