```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge, RidgeCV, Lasso, LassoCV,LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
np.random.seed(0)
```

```python
df = pd.read_csv('Boston_Data.csv')
df
```

```python
df.info()
```

```python
y = df[['crim']]
X = df.drop(['crim'],axis=1).astype('float64')
X
```

Scaling

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test , y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=0)


X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

# Initial Scores - Linear Regressıon

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
prediction = lr.predict(X_test)
test_score = mean_squared_error(y_test,prediction)

prediction_train = lr.predict(X_train)
train_score = -cross_val_score(lr,X_train,y_train,cv=5,scoring='neg_mean_squared_error').mean()
                                
print("train mse : ",train_score, "test mse : ",test_score)
```

# Subset Selection

# Exhaustive Search

```python
from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS
# Create Logistic Regression classifier
lr = LinearRegression()
efs = EFS(lr, 
           min_features=1,
           max_features=13,
           scoring='neg_mean_squared_error',
           print_progress=True,
           cv=5)

efs.fit(X_train, y_train)

print('Best MSE score: %.2f' % efs.best_score_ * (-1))
print('Best subset:', efs.best_idx_)
```

#   Values are neg_mean_square, the one with abs value smaller, better

```python
efs_analysis = pd.DataFrame.from_dict(efs.get_metric_dict()).T
efs_analysis.sort_values(by = ['avg_score'],ascending = False)
```

Exhaustive Search Does not give much idea, as scores are so close, it basically tries every option

```python
X_train_subset1 = X_train[:,[0,1,3,6,7,9,11,12]]
X_test_subset1 = X_test[:,[0,1,3,6,7,9,11,12]]
```

```python
lr = LinearRegression()
lr.fit(X_train_subset1,y_train)
prediction = lr.predict(X_test_subset1)
test_score = mean_squared_error(y_test,prediction)

train_score = -cross_val_score(lr,X_train_subset1,y_train,cv=5,scoring='neg_mean_squared_error').mean()
                                
print("train mse : ",train_score, "test mse : ",test_score)
```

# Sequential Forward

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
lr = LinearRegression()
sfs = SFS(lr,
           k_features=X_train.shape[1],
           forward=True, 
           floating=False, 
           verbose=2,
           scoring='neg_mean_squared_error',
           cv=5)

sfs = sfs.fit(X_train, y_train)
sfs.subsets_
```

```python
sfs_an1 = pd.DataFrame.from_dict(sfs.get_metric_dict()).T
sfs_an1.sort_values(by = ['avg_score'],ascending = False)
```

# RIDGE 

```python
alphas = 10**np.linspace(5,-2,100)*0.5
alphas
```

At previous stage, I have already scaled the Data.

```python
ridgecv = RidgeCV(alphas = alphas,cv = 5, scoring = 'neg_mean_squared_error')
ridgecv.fit(X_train, y_train)
ridgecv.alpha_
```

```python
ridge5 = Ridge(alpha = ridgecv.alpha_)
ridge5.fit(X_train, y_train)
test_score = mean_squared_error(y_test, ridge5.predict(X_test))
train_score = -cross_val_score(ridge5,X_train,y_train,cv=5,scoring='neg_mean_squared_error').mean()

print("train mse : ",train_score, " test mse : ",test_score)
```
