```
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2) + np.cos(x)
    
#Creates just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
ax.set_xlabel("x values")
```

```
f, (ax1, ax2) = plt.subplots(1, 2, 
                             sharey=True)
derivative = 2 * x * np.cos(x**2) - np.sin(x)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.plot(x, derivative)

f.set_size_inches(4.5, 0.5)
plt.show()
```

```

import matplotlib.pyplot as plt

rows, cols = 2, 3
fig, ax = plt.subplots(rows, cols,
                       sharex=True, 
                       sharey=False, figsize = (cols*2,rows*2))

for row in range(rows):
    for col in range(cols):
        ax[row, col].text(0.5, 0.5, 
                          str((row, col)),
                          color="green",
                          fontsize=18, 
                          ha='center')

#fig.set_size_inches(12,2)
plt.show()

```

```
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
sns.set() # Setting seaborn as default style even if use only matplotlib
titanic = sns.load_dataset("titanic")
```

```
categorical_vars = ['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch',
       'embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town',
       'alive', 'alone']

fig, axs = plt.subplots(nrows=2, ncols=2)

for i, var in enumerate(categorical_vars):
    row = i//2
    pos = i % 2    
    plot = sns.countplot(x=var, data=titanic, ax=axs[row][pos])
```

```
categorical_vars = ['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch',
       'embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town',
       'alive', 'alone']
num_plots = len(titanic.columns)
total_cols = 2
total_rows = num_plots//total_cols + 1
fig, axs = plt.subplots(nrows=total_rows, ncols=total_cols,
                        figsize=(3*total_cols, 3*total_rows), constrained_layout=True)

for i, var in enumerate(titanic.columns):
    row = i//total_cols
    pos = i % total_cols    
    sns.countplot(x=var, data=titanic, ax=axs[row][pos])

plt.show()
```
