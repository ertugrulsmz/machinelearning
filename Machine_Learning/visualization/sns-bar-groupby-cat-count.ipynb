```python
import seaborn as sns
import pandas as pd

planets = sns.load_dataset("diamonds")
df = planets.copy()
df.head(), df.shape
```

### catplot -distrubution of prices by categories (min-max)

```python
sns.catplot(x="cut", y="price", data=df)
```

```python
print("max of price : ",df.price.max())
print("max of categories : ")
df.groupby("cut")['price'].max()
```

# bar plot by categories --mean

```python
sns.barplot(x='cut',y='price', hue="color", data=df)
```

```python
df.groupby(['cut','color'])['price'].mean()
```

```python
sns.countplot(x='cut',hue='color', data=df)
```

```python
sns.countplot(x='cut',data=df)
```
