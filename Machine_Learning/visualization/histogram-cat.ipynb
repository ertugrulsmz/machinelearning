```python
import seaborn as sns
import pandas as pd

diaomonds = sns.load_dataset("diamonds")
df = diaomonds.copy()

df.head()
```

### histogram

```python
sns.displot(df.price, kde = False)
```

```python
df.price.describe()
```

```python
sns.displot(df.price, bins = 3)
```

```python
sns.distplot(df.price)
```

```python
sns.distplot(df.price, hist=False)
```

```python
sns.kdeplot("price",data=df, shade=True)
```

### histogram by category - cat by category

```python
(
    sns
    .FacetGrid(df, hue="cut", height=5, xlim=(0,10000))
    .map(sns.kdeplot, "price", shade = True)
    .add_legend()
)
```

ideal (the best) is suprisingly seem cheapest in distrubution

```python
sns.catplot(x="cut", y="price",hue="color", kind="point", data=df)
```
