```python
import seaborn as sns
tips = sns.load_dataset("tips")

df = tips.copy()
df.head()
```

#### scatter plot

#### 2 continous variable together

```python
sns.scatterplot(x="total_bill",y="tip", data=df)
```

### crosses

```python
sns.scatterplot(x="total_bill", y="tip", hue="time", data=df)
```

```python
sns.scatterplot(x="total_bill", y="tip", size="size", data=df)
```

```python
sns.scatterplot(x="total_bill", y="tip", size="size", hue="size", data=df)
```

### lm plot

```python
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x="total_bill", y="tip", data=df)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=df)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="day", data=df)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", row="sex", data=df)
```

# pair plot

```python
sns.pairplot(df, hue="time")
```
